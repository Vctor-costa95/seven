```python
def remove_jpg_files(folder_path):
    try:
        # List all files in the specified folder
        files = os.listdir(folder_path)

        # Filter and remove JPG files
        for file in files:
            if file.lower().endswith(".jpg"):
                file_path = os.path.join(folder_path, file)
                os.remove(file_path)
                print(f"Removed: {file_path}")

        print("JPG files removed successfully.")
    
    except Exception as e:
        print(f"Error: {e}")

# Replace 'your_folder_path' with the actual path of the folder containing the JPG files
folder_path = r"C:\Users\jvict\Documents\seven\seven"
remove_jpg_files(folder_path)
```
